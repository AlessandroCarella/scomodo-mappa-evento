```python
import json
```

```python
with open ("storie.json") as f:
    data = json.load(f)
```

```python
citta = set()
for elem in data:
    citta.add(elem["cittaPartenza"])
    citta.add(elem["cittaArrivo"])
```

```python
citta
```

```python
import urllib.request
import json
import sys
import time

def get_coordinates(city: str) -> tuple[float, float]:
    city_encoded = urllib.parse.quote(city)
    url = f"https://photon.komoot.io/api/?q={city_encoded}&limit=1"
    req = urllib.request.Request(url, headers={"User-Agent": "geo-lookup-script/1.0"})
    with urllib.request.urlopen(req) as response:
        data = json.loads(response.read())
    if not data["features"]:
        raise ValueError(f"City '{city}' not found.")
    lng, lat = data["features"][0]["geometry"]["coordinates"]
    return lat, lng

output = []
for city in citta:
    lat, lng = get_coordinates(city)
    output.append({"name": city, "lat": lat, "lng": lng})
    print(f"✓ {city}: {lat}, {lng}")
    time.sleep(2)  # be conservative with the delay
```

```python
with open ("Locations2.json", "w", encoding="utf-8") as f:
    json.dump(output, f, indent=4, ensure_ascii=False)
```
